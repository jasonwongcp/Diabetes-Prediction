## Data Cleaning

The bank has provided us with this snapshot of demographic and campaign data related to the term deposit product. First, let's read in the data and take a look at what we're working with, and assess the level of cleaning and preprocessing that needs to be done.

### Duplicate data

We assumed that each row in the data corresponds to a unique patient.

```python
raw_data.duplicated()
```

```python
raw_data.duplicated().sum()
```

The number of duplicate rows is 6672. Let's check the percentage of data that is duplicated.

```python
raw_data.duplicated().sum()/raw_data.shape[0]*100
```

Given that the goal of my analysis is to predict whether a patient has diabetes, it is essential to carefully consider the presence of duplicate rows and their potential impact on my predictive model. In this case, having nearly 10% duplicate rows could be significant, and dropping them might be advisable for 3 reasons. 

1. Model Performance: Duplicate rows can affect the performance of my predictive model. When duplicate rows are present in the dataset, they can introduce bias and artificially inflate the importance of certain observations. This can lead to overfitting and poor generalization of the model to new data. Removing duplicate rows can help mitigate these issues and improve the model's performance.

2. Data Integrity: Duplicated rows may create noise and inconsistencies in the dataset, which can affect the accuracy and reliability of the analysis. By removing the duplicate rows, I ensure a cleaner and more reliable dataset for my modeling process.

3. Computational Efficiency: Duplicates can increase the computational burden during model training, especially if I have a large dataset. Removing duplicate rows can improve computational efficiency by reducing redundant computations and memory usage.

Considering the potential benefits in terms of model performance, data integrity, and computational efficiency, I have decided to drop duplicate rows, especially when they constitute a significant proportion of the dataset.

```python
df_deduplicated = raw_data.drop_duplicates()
```

### 2. Distribution Plots

•	Create distribution plots for each feature to see their individual distributions.

```python
#check frequency of all values in the column

# All data columns except for color
feature_cols = [x for x in df_deduplicated.columns]
plt.figure(figsize=(25, 35))

# loop for subplots
for i in range(len(feature_cols)):
    plt.subplot(8, 5, i + 1)
    sns.countplot(x=feature_cols[i], data=df_deduplicated, palette='Set1')  # Use countplot instead of hist
    plt.title(feature_cols[i])
    plt.xticks(rotation=90)
    plt.xlabel(None)
    plt.ylabel('Frequency')
    plt.subplots_adjust(hspace=0.5)  # Add vertical spacing between subplots

plt.tight_layout()
plt.show()
```

```python
# Calculate the total number of data points (rows) in the DataFrame
total_data_points = df_deduplicated.shape[0]

# Function to display percentages on the count plot
def show_percentage_on_barplot(ax):
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(f'{height / total_data_points * 100:.2f}%', 
                    xy=(p.get_x() + p.get_width() / 2., height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

# Set a custom color palette for the plots
custom_palette = sns.color_palette("Set2")

# Create a 1x3 subplot layout
fig, axes = plt.subplots(1, 3, figsize=(15, 6))

# Check the frequency of categories for 'CholCheck' column
sns.countplot(x='CholCheck', data=df_deduplicated, ax=axes[0], palette=custom_palette)
axes[0].set_title('Frequency of Categories in CholCheck Column', fontsize=12)
axes[0].set_xlabel('CholCheck Category', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)

# Display percentages on the plot
show_percentage_on_barplot(axes[0])

# Check the frequency of categories for 'HvyAlcoholConsump' column
sns.countplot(x='HvyAlcoholConsump', data=df_deduplicated, ax=axes[1], palette=custom_palette)
axes[1].set_title('Frequency of Categories in HvyAlcoholConsump Column', fontsize=12)
axes[1].set_xlabel('HvyAlcoholConsump Category', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)

# Display percentages on the plot
show_percentage_on_barplot(axes[1])

# Check the frequency of categories for 'Stroke' column
sns.countplot(x='Stroke', data=df_deduplicated, ax=axes[2], palette=custom_palette)
axes[2].set_title('Frequency of Categories in Stroke Column', fontsize=12)
axes[2].set_xlabel('Stroke Category', fontsize=12)
axes[2].set_ylabel('Frequency', fontsize=12)

# Display percentages on the plot
show_percentage_on_barplot(axes[2])

# Adjust the space between subplots
plt.tight_layout()

# Add gridlines to the plots
for ax in axes:
    ax.grid(axis='y', linestyle='--', alpha=0.7)

# Remove the top and right spines for cleaner appearance
for ax in axes:
    sns.despine(ax=ax, right=True, top=True)

# Adjust the font size of the tick labels
for ax in axes:
    ax.tick_params(axis='both', which='major', labelsize=10)

plt.show()
```

### 3. Numerical Features and Categorical features V.S. Target

•	Create scatter plots or box plots to visualize the relationship between each numerical feature (columns 0-16) and the target 'Diabetes'. This can help identify potential patterns or correlations.

```python
# Selecting the numeric columns for box plots
numeric_cols = ['Age', 'BMI', 'MentHlth', 'PhysHlth','GenHlth']

# Box plots for numerical features vs. Diabetes
fig, axes = plt.subplots(1, len(numeric_cols), figsize=(20, 5))
for i, col in enumerate(numeric_cols):
    sns.boxplot(x='Diabetes', y=col, data=df_deduplicated, ax=axes[i])
    axes[i].set_title(f'{col} by Diabetes')
    axes[i].set_xlabel('Diabetes')
    axes[i].set_ylabel(col)

plt.tight_layout()
plt.show()

# Bar plots for binary categorical features vs. Diabetes
binary_categorical_cols = ['Sex', 'HighChol', 'CholCheck', 'Smoker', 'HeartDiseaseorAttack', 'PhysActivity','Fruits','Veggies','HvyAlcoholConsump','DiffWalk','Stroke','HighBP']

fig, axes = plt.subplots(2, 6, figsize=(18, 8))
for i, col in enumerate(binary_categorical_cols):
    row, col_index = divmod(i, 6)  # Calculate the row and column for the subplot
    sns.barplot(x=col, y='Diabetes', data=df_deduplicated, ax=axes[row, col_index])
    axes[row, col_index].set_title(f'{col} vs. Diabetes')
    axes[row, col_index].set_xlabel(col)
    axes[row, col_index].set_ylabel('Probability of Diabetes')

plt.tight_layout()
plt.show()
```

Let’s take a look at physical features. The probability of diabetes is significantly higher for patients with high cholesterol and high blood pressure, suggesting these two are important contributing factors to the development of diabetes. The probability of diabetes does not increase when the patients consume high amount of alcohol or are smokers, suggesting that high intake of alcohol or smoking does not increase the risks of diabetes development.

### 4.Pairwise Correlation Heatmap

•	Generate a heatmap to visualize the pairwise correlation between numerical features using Seaborn's sns.heatmap().

```python
# Compute pairwise correlation between numerical features
correlation_matrix = df_deduplicated[numeric_cols].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Pairwise Correlation Heatmap of Numerical Features')
plt.show()
```

### 5.Correlation with Target

•	Compute the correlation of each feature with the target variable 'Diabetes' and visualize it using a bar plot.

```python

# Compute correlation of each feature with the target variable 'Diabetes'
correlation_with_target = df_deduplicated.corr()['Diabetes'].drop('Diabetes')

# Plot the bar plot using Seaborn with a fancier color
plt.figure(figsize=(10, 6))
custom_color = sns.color_palette("pastel")  # Choose any color palette you like
ax = sns.barplot(x=correlation_with_target.index, y=correlation_with_target.values, palette=custom_color, linewidth=2.5)

# Add a horizontal line at 0.3
plt.axhline(y=0.3, color='r', linestyle='--', label='Correlation >= 0.3')

# Add labels to the bars that are higher than 0.3
for i, v in enumerate(correlation_with_target):
    if v >= 0.3:
        ax.text(i, v + 0.003, f"{v:.2f}", ha='center', va='bottom', color='green')

plt.xlabel('Features')
plt.ylabel('Correlation with Diabetes')
plt.title('Correlation of Features with Diabetes')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

It shows that General Health adn High Blood Pressure show stronger linear relationship with Diabetes. It is worth pointing out that even though High cholesterol significantly increases the the risk of developing diabetes, the correlation with diabetes is low. It suggests that the linear relatioships are not strong and it is possible that there may be some other confounding factors.
